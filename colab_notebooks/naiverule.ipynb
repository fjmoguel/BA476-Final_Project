```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.metrics import mean_squared_error


# Load datasets
df_01 = pd.read_csv('/content/drive/My Drive/flight_data_regression_with_features.csv')
df_15 = pd.read_csv('/content/drive/My Drive/flight_data_classification_with_features.csv')
```

```python
# Function to calculate mean and MSE
def naive_rule_mse(df, target_column):
    # Calculate the mean of the target column
    mean_value = df[target_column].mean()
    # Predict the mean for all rows
    predictions = [mean_value] * len(df)
    # Calculate the MSE
    mse = mean_squared_error(df[target_column], predictions)
    return mean_value, mse

# Apply to both datasets
mean_01, mse_01 = naive_rule_mse(df_01, "ARR_DELAY")
mean_15, mse_15 = naive_rule_mse(df_15, "ARR_DELAY")

# Print results
print("Dataset 0-1:")
print(f"Mean ARR_DELAY: {mean_01}")
print(f"MSE: {mse_01}")

print("\nDataset 15+:")
print(f"Mean ARR_DELAY: {mean_15}")
print(f"MSE: {mse_15}")
```
